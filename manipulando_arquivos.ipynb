- Usando o bloco with para manipular arquivos
- Usando o módulo os (Operational System) para manipular arquivos

````python
import os
from datetime import datetime
import time

data_atual = datetime.now()
nome_arquivo = 'arquivo_teste.txt'

print(f"Cria o arquivo {nome_arquivo}")
with open("arquivo_teste.txt", mode='w') as file:
  file.write(f"Conteúdo do arquivo teste em {data_atual}")

print(f"Arquivo {nome_arquivo} criado com sucesso!")
````

# Resumos dos principais comandos para manipular arquivos usando o Linux

Lista dos **principais comandos básicos para manipular arquivos no Linux**, essenciais para gerenciamento e operações do dia a dia:

---

### **1. Navegação e Visualização**
| Comando             | Descrição                                                                 |
|---------------------|---------------------------------------------------------------------------|
| `pwd`               | Mostra o diretório atual (*Print Working Directory*).                     |
| `ls`                | Lista arquivos e pastas no diretório atual.                               |
| `ls -l`             | Lista com detalhes (permissões, tamanho, data).                           |
| `ls -a`             | Lista **todos** os arquivos (incluindo ocultos, que começam com `.`).     |
| `tree`              | Mostra a estrutura de diretórios em forma de árvore (instale com `sudo apt install tree`). |

---

### **2. Criação e Remoção**
| Comando             | Descrição                                                                 |
|---------------------|---------------------------------------------------------------------------|
| `touch arquivo.txt` | Cria um arquivo vazio (ou atualiza a data de modificação se existir).     |
| `mkdir pasta`       | Cria uma nova pasta.                                                      |
| `mkdir -p pasta/subpasta` | Cria uma estrutura de pastas (incluindo subpastas se não existirem). |
| `rm arquivo.txt`    | **Remove** um arquivo (*não vai para a lixeira!*).                        |
| `rm -r pasta`       | Remove uma pasta **e todo seu conteúdo** (*cuidado!*).                    |
| `rmdir pasta`       | Remove uma pasta **apenas se estiver vazia**.                             |

---

### **3. Copiar, Mover e Renomear**
| Comando                          | Descrição                                                                 |
|----------------------------------|---------------------------------------------------------------------------|
| `cp origem.txt destino.txt`      | Copia um arquivo.                                                         |
| `cp -r pasta_origem pasta_destino` | Copia uma pasta **recursivamente** (com todo o conteúdo).              |
| `mv origem.txt destino.txt`      | Move ou **renomeia** um arquivo/pasta.                                    |
| `mv arquivo.txt /outro/caminho/` | Move o arquivo para outro diretório.                                      |

---

### **4. Visualização e Edição**
| Comando             | Descrição                                                                 |
|---------------------|---------------------------------------------------------------------------|
| `cat arquivo.txt`   | Mostra o conteúdo do arquivo no terminal.                                 |
| `less arquivo.txt`  | Visualiza o arquivo com scroll (saia pressionando `q`).                   |
| `head arquivo.txt`  | Mostra as **10 primeiras linhas** do arquivo.                             |
| `tail arquivo.txt`  | Mostra as **10 últimas linhas** (útil para logs).                         |
| `nano arquivo.txt`  | Abre o arquivo no editor **Nano** (simples).                              |
| `vim arquivo.txt`   | Abre no editor **Vim** (avançado).                                       |

---

### **5. Permissões e Propriedades**
| Comando                     | Descrição                                                                 |
|-----------------------------|---------------------------------------------------------------------------|
| `chmod 755 arquivo.txt`     | Altera permissões (leitura/escrita/execução).                            |
| `chown usuario:grupo arquivo.txt` | Muda o dono e grupo do arquivo.                                     |
| `stat arquivo.txt`          | Exibe detalhes como tamanho, datas de acesso e permissões.               |

---

### **6. Busca e Comparação**
| Comando                          | Descrição                                                                 |
|----------------------------------|---------------------------------------------------------------------------|
| `find /caminho -name "*.txt"`    | Busca arquivos pelo nome (ex: todos `.txt`).                              |
| `grep "texto" arquivo.txt`       | Procura por um **texto** dentro do arquivo.                               |
| `diff arquivo1.txt arquivo2.txt` | Compara dois arquivos e mostra as diferenças.                             |

---

### **7. Compactação e Descompactação**
| Comando                                      | Descrição                                                                 |
|----------------------------------------------|---------------------------------------------------------------------------|
| `tar -cvf arquivo.tar pasta/`               | Cria um arquivo `.tar` (compactação sem compressão).                      |
| `tar -xvf arquivo.tar`                      | Extrai um `.tar`.                                                         |
| `gzip arquivo.txt`                          | Compacta para `.gz` (original é **substituído**).                         |
| `gunzip arquivo.txt.gz`                     | Descompacta `.gz`.                                                        |
| `zip arquivo.zip pasta/`                    | Cria um `.zip` (instale com `sudo apt install zip`).                      |
| `unzip arquivo.zip`                         | Extrai um `.zip`.                                                         |

---

### **Exemplos Práticos**
1. **Criar e copiar um arquivo**:
   ```bash
   touch teste.txt
   cp teste.txt backup.txt
   ```

2. **Buscar todos os arquivos `.log`**:
   ```bash
   find /var/log -name "*.log"
   ```

3. **Verificar permissões**:
   ```bash
   ls -l /etc/passwd
   ```

---

### **Avisos Importantes**
- **`rm -r` e `rm` não têm lixeira**: Os arquivos são **excluídos permanentemente**.
- Use `sudo` antes de comandos que requerem permissões de root (ex: `sudo rm`).
